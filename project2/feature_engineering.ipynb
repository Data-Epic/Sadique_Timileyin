```python
import pandas as pd
import gspread as gs
from sklearn.preprocessing import OrdinalEncoder
import numpy as np
```

```python
gc = gs.service_account('my-project.json')
sh = gc.open_by_url('https://docs.google.com/spreadsheets/d/1SUlcukpgxf6pfFQbj6DKQZXSlOhyBOSuBdbmQ2ZbhTM/edit#gid=327494290')
worksheet = sh.worksheet('missing_data')
dataframe = pd.DataFrame(worksheet.get_all_records())
df = pd.DataFrame(dataframe)
```

```python
df = df.replace('', np.nan)
```

```python
df.info()
```

```python
df.drop(columns = ['TIDF Compliance', 'Voluntary Soft-Story Retrofit'], inplace = True)
```

```python
df['Site Permit'] = df['Site Permit'].fillna(df['Site Permit'].mode()[0])
```

```python
df['Structural Notification'] = df['Structural Notification'].fillna(df['Structural Notification'].mode()[0])
```

```python
df['Number of Existing Stories'] = df['Number of Existing Stories'].fillna(df['Number of Existing Stories'].mean()).round()
```

```python
df['Number of Proposed Stories'] = df['Number of Proposed Stories'].fillna(df['Number of Proposed Stories'].mean()).round()
```

```python
df['Fire Only Permit'] = df['Fire Only Permit'].fillna(df['Fire Only Permit'].mode()[0])
```

```python
df['Street Suffix'] = df['Street Suffix'].fillna(df['Street Suffix'].mode()[0])
```

```python
df.drop(columns = ['Unit', 'Unit Suffix'], inplace = True)
```

```python
df['Estimated Cost'] = df['Estimated Cost'].fillna(df['Estimated Cost'].mean()).round()
df['Estimated Cost'] = df['Estimated Cost'].replace(1, df['Estimated Cost'].mean().round())
```

```python
df['Revised Cost'] = df['Revised Cost'].fillna(df['Revised Cost'].mean()).round()
```

```python
# df['Revised Cost'] = df['Revised Cost'].replace(2, df['Revised Cost'].mean().round())
```

```python
df.info()
```

```python
df.drop(columns='Street Number Suffix', inplace= True)
```

```python
df['Issued Date'] = pd.to_datetime(df['Issued Date'])

df['First Construction Document Date'] = pd.to_datetime(df['First Construction Document Date'])

df['Completed Date'] =pd.to_datetime(df['Completed Date'])

df['Filed Date'] = pd.to_datetime(df['Filed Date'])

df['Current Status Date'] = pd.to_datetime(df['Current Status Date'])

df['Permit Creation Date'] = pd.to_datetime(df['Permit Creation Date'])

df['Permit Expiration Date'] = pd.to_datetime(df['Permit Expiration Date'])
```

```python
df[['Latitude', 'Longitude']] = df['Location'].str.split(',', 2, expand=True)
```

```python
df['Latitude'] = df['Latitude'].str.replace('(', '')
```

```python
df['Longitude'] = df['Longitude'].str.replace(')', '')
```

```python
df.drop(columns='Location', inplace = True)
```

```python
df['Zipcode'] = df['Zipcode'].fillna('UNKNOWN')
```

```python
df['Neighborhoods - Analysis Boundaries'] = df['Neighborhoods - Analysis Boundaries'].fillna('UNKNOWN')
```

```python
df['Supervisor District'] = df['Supervisor District'].fillna(df['Supervisor District'].mean().round())
```

```python
df['Proposed Construction Type'] = df['Proposed Construction Type'].fillna(df['Proposed Construction Type'].mean()).round()
```

```python
df.drop(columns =['Permit Type', 'Permit Type Definition', 'Block', 'Lot', 'Street Number', 'Street Name', 'Street Suffix', 'Description', 'Current Status', 'Current Status Date', 'Filed Date', 'Issued Date', 'Completed Date', 'First Construction Document Date', 'Structural Notification', 'Number of Existing Stories', 'Number of Proposed Stories', 'Fire Only Permit', 'Estimated Cost', 'Existing Use', 'Existing Units', 'Proposed Use', 'Proposed Units', 'Plansets', 'Existing Construction Type Description', 'Proposed Construction Type', 'Proposed Construction Type Description', 'Site Permit', 'Supervisor District', 'Neighborhoods - Analysis Boundaries', 'Latitude', 'Longitude', 'Permit Expiration Date', 'Record ID', 'Permit Number', 'Existing Construction Type'], inplace = True)
```

```python
df['Revised_Cost_Range'] = pd.cut(df['Revised Cost'], bins = [-np.inf, 1000, 10000, 100000, 1000000, 10000000, np.inf],
                            labels=['Very Low', 'Low', 'Medium', 'High', 'Very High', 'Extremely High'])
```

```python
for x in df['Revised_Cost_Range'].unique():
    df[x] = np.where(df['Revised_Cost_Range']==x, 1, 0)
```
